```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

def generate_email(sentiment):
    model_name = 'gpt2'
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)
    model = GPT2LMHeadModel.from_pretrained(model_name)

    if sentiment == 'positive':
        prompt = 'I wanted to say thank you for the great service that your bank is offering for my bank account.'
    else:
        prompt = 'I am unbelievable angry about your bank and I have a complaint.'
        
    inputs = tokenizer.encode(prompt, return_tensors='pt')
    outputs = model.generate(inputs, max_length=100, num_return_sequences=1, do_sample=True, no_repeat_ngram_size=2, temperature=0.9)
    email_body = tokenizer.decode(outputs[0], skip_special_tokens=True)

    email_body = email_body.replace(",", " ").replace("\n", " ")
    print(email_body)
    return email_body
```

```python
positive_emails = [generate_email('positive') for _ in range(0, 10)]
negative_emails = [generate_email('negative') for _ in range(0, 10)]
```

```python
import random
from faker import Faker
import datetime

seed = 18102022

random.seed(seed)
Faker.seed(seed)

fake = Faker()

negative_subjects = ['Urgent question', 'Please help me', 'Complaint!!!']
positive_subjects = ['Thanks!', 'Great bank', 'You are awesome']

#num_customers_to_generate = 1000000
num_customers_to_generate = 100000
num_mails_per_customer = 10

with open(f'../data/synthetic_customers_{num_customers_to_generate}.csv', 'w') as customers_file:
    with open(f'../data/synthetic_mails_{num_customers_to_generate}.csv', 'w') as mails_file:
        for customer_id in range(0, num_customers_to_generate):
            
            customer_email = fake.ascii_company_email()
            
            if random.random() > 0.25:
                country = random.choice(['UK', 'DE', 'FR'])
            else:        
                country = fake.country_code()
        
            bank = fake.company().replace(", ", " ")
        
            if random.random() > 0.9:
                level = 'premium'
            else:        
                level = 'standard'
            
            customers_file.write(f'{customer_id},{customer_email},{bank},{country},{level}\n')

            
            for _ in range(0, num_mails_per_customer):
                mail_id = fake.uuid4()
                raw_date = fake.date_between(datetime.date(2022, 8, 1), datetime.date(2024, 6, 15))

                if random.random() > 0.6:
                    subject = random.choice(negative_subjects) 
                    email = random.choice(negative_emails)

                    if country == 'DE' and random.random() > 0.99:
                        subject += ' My credit card is ' + fake.credit_card_number()
                        print(subject)
                
                else:        
                    subject = random.choice(positive_subjects)
                    email = random.choice(positive_emails)

                mails_file.write(f'{mail_id},{customer_email},{raw_date},{subject},{email}\n')
```
